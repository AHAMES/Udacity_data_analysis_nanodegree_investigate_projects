```python
movies['director'] = np.where(movies['director'].str.find("|") != -1, "C", movies['director'])
movies['director'] = np.where(movies['director'] != "C" , "S", movies['director'])
```

```python
movies = movies[movies.revenue>10000]
movies = movies[movies.budget>10000]
```

Multiple movies from different genres were made over the years, some were movies had several genres built into them, but which genres were conistently the more successful over the years? Also does popularity improve the revenue or not?

```python
# Continue to explore the data to address your additional research
#   questions. Add more headers as needed if you have more questions to
#   investigate
fig=plt.figure(figsize=(20,10))
ax=fig.add_axes([0,0,1,1])
np.random.seed(0)
for genre in genres.columns:
    genre_movies = reducedMovies[reducedMovies[genre] == 1]
    revenue = genre_movies.revenue
    time_range = genre_movies.release_year
    
    ax.scatter(time_range, revenue,label=genre)
    ax.set_xlabel('Years')
    ax.set_ylabel('Revenue')
    ax.set_title('scatter plot')
plt.legend(loc='best')
plt.show()
```

<b>The following charts are meant to provide an understanding on how much each genre has as share in the respective realm </b>

This one for example shows documentary to the real lowest in terms of revenue, this is likely because no one watches a documentary in theatre.<br>
Horror on average also has very little revenue, likely by the fact that it is R rated usually, and that most horror movies low budget generally terrible movies

```python
def drawPieChart(feature):
    revenues_values = []
    genres_labels= []
    fig = plt.figure(figsize=(7,7))
    ax = fig.add_axes([0,0,1,1])
    ax.axis('equal')
    for genre in genres.columns:
        genre_movies = reducedMovies[reducedMovies[genre] == 1]
        mean_rev = genre_movies[feature].mean()
        if mean_rev==0 or math.isnan(mean_rev):
            continue

        genres_labels.append(genre[:genre.find("genre")-1])
        revenues_values.append(mean_rev)

    revenues_values, genres_labels = (list(t) for t in zip(*sorted(zip( revenues_values,genres_labels))))
    explode = np.full(shape=len(revenues_values),fill_value=0.05)    
    ax.pie(revenues_values, labels = genres_labels, autopct='%.1f%%', explode = explode)

    ax.set_title('Pie chart')
    plt.title("Best genres by average "+feature)
    plt.show()
```

```python
drawPieChart("revenue")
```

This pie chart here shows a decorrelation between it and the previous pie chart, while science fiction is the most popular, it is not by default the most earning.

Horror movies as well generally earn more than TV movies, which could be simply explained that again, no one watches a TV movie in theatre, because they are on TV.

Most genres are generally out of place an unalligned when comparing their revenue to popularity, generally showing that movies may not earn more, but they can be appreciated for what they are later on.

```python
drawPieChart("popularity")
```

The next two charts tell an interesting story, as documentary is still getting low score on vote counts, meaning that not many vote for documentaries.

However in the next chart we find that while most gernes are generally equal in average votes, the highest one by far are documentaries, this indicates that while very little of the documentaries receive any vote at all, those votes are always highly appreciative, and it shows that whoever is watching documentaries is always the type that appreciates any quality provided knowledge

```python
drawPieChart("vote_count")
```

```python
drawPieChart("vote_average")
```

<b>While the previous charts indicate audience reception. It does not indicate any number, how many of each genre is made, and therefore which ones are more preferred by producers companies. The following chart tries to find the answer to that<b>

From this bar chart, I have concluded that westerns are underated, sure they don't earn a lot, but they are constantly proven to be popular and have a high view rating (if votes are to be believed as views). While dramas, comedies are overated considering they are the most while not delivering on most scales, being some of the lowest in every given standard.

```python
fig = plt.figure(figsize=(17,7))
genres_labels = []
genre_counts = []
ax = fig.add_axes([0,0,1,1])
for genre in genres.columns:
    genre_movies = reducedMovies[reducedMovies[genre] == 1]
    count = genre_movies.vote_average.count()
    if count==0 or math.isnan(count):
        continue
    
    genres_labels.append(genre[:genre.find("genre")-1])
    genre_counts.append(count)
    
ax.barh(genres_labels,genre_counts)
plt.show()
```
